```python
# parse_dates=["day"] --> untuk mengub
# NB: 2 baris code berikut harus berada dalam satu sell, kalau engga maka akan error
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv", parse_dates=["day"])
df.set_index(df["day"], inplace=True) # Untuk membuat kolom day enter satu kali dalam output dataframe
df
```

```python
new_df = df.fillna(0) # untuk mengganti nilai NaN ke angka "0"
new_df
```

```python
# Metode berikut untuk mengubah nilai NaN untuk kolom tertentu
# Untuk mengubah dataframe yg tipe datanya bercampur integer dan string maka menggunakan cara ini , lalu simpan didataframe baru kita
# NB: untuk mengubah nilai string yg NaN maka sesuaikan dg maksud variable atau maksud dari isi variablenya
new_df = df.fillna({
    "temperature" : 0,
    "windspeed" : 0,
    "event" : "no event"
})
new_df
```

```python
# Bisa juga mengubah data yg NaN ke data sebelumnya (tapi ingat yg mana yg lebih cocok untuk tujuan analisis kita)
new_df = df.fillna(method="ffill")
new_df
```

```python
# Bisa juga mengganti nilai NaN ke nilai setelahnya
new_df = df.fillna(method="bfill")
new_df
```

```python
# adalah untuk menghapus baris yang memiliki seluruh nilai yang kosong (NaN) dalam sebuah dataset. Dengan
# menggunakan paraeter how="all", hanya baris yang memiliki seluruh nilai yang kosong akan dihapus
new_df = df.dropna(how="all")
new_df
```

```python
# This is method to delete the NaN value if a NaN value there is a row one
new_df = df.dropna(thresh=1)
new_df
```

```python
new_df2 = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/6_handling_missing_data_replace/weather_data.csv")
new_df2
```

```python
import numpy as np
# Replace/changed the value wich not suitable with values other in every columns

new_df2 = new_df2.replace([-99999, -88888, 0], np.NaN)
new_df2
```

```python
# We also can use this a method but result/output same just

# NB: angka ini sebenarnya tidak ada di dalam data set, tapi angka tersebut dimodivikasi pada data set, modivikasinya saat video youtube yg sedang berlangsung
new_df2 = new_df2.replace({
    "temperature" : -99999,
    "windspeed" : -99999,
    "event" : "0"
  }, np.NaN)
new_df2
```

```python
# saran saja metode ini jangan digunakan, gunakanlah metode setelah ini
# Menghapus nilai2 "pada setiap kolom" yang mengandung nilai alfabet/huruf

# Coba tanya ke chatgpt: Bagaimana cara menghapus nilai2 alfabet/huruf pada kolom tertentu
new_df2 = new_df2.replace("[A-Za-z]" , " ", regex=True)
new_df2
```

```python
# Menghapus nilai string yang ada dalam kolom integer (metode ini untuk kolom2 tertentu)
new_df2 = new_df2.replace({
    "temperature" : "[A-Za-z]",
    "windspeed" : "[A-Za-z]"
  }, " ", regex=True)
new_df2
```
