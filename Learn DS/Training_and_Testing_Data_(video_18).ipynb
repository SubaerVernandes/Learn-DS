# Prepare data train and data test

```python
x = df[["Mileage", "Age(yrs)"]] # variable dependent
y = df["Sell Price($)"]         # variable independent
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Window ini jika dirunning ulang maka outputnya akan berubah setiap dirunning pada output window ini
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2) # 20% data latih 80% data uji
x_train

# NB: rest_size=0.3 elseif 30% data latih 70% data uji
```

```python
# Jika code sebelumnya dijalankan berulang2 mendapat output yg berbeda maka kita use random_state=parameter for get ouput wich/whose statitation/statis/tetap on every running two codes it
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
x_train
```

```python
# Return to code whose/wich earlier because in tutorial not using random_state parameter
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
x_train
```

# Applied the model

```python
from sklearn.linear_model import LinearRegression
clf= LinearRegression()
```

```python
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
```
