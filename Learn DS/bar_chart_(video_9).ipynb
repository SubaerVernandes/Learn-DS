```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company= ["GOOGL", "AMZN", "MSFT", "FB"]
revenue = [90, 136, 89, 27]
```

```python
# Untuk membuat barchat maka kita tidakbisa menjadikan variabel list company sebagai sumbu x, maka kita hitung dulu menggunakan numpy lalu jadikanlah itu sebagai sumbu x
xpos = np.arange(len(company))
xpos
```

```python
# creating the graph righ now use bar_chart

# ypos --> untuk ukuran grafik dg deret setengah-setengah sesuai indeks
# revenue adalah barnya
# sumbu y akan otomatis ada sendiri nilainya hal ini sesuai tinggi bar atau nilai
plt.bar(xpos, revenue)
```

```python
plt.bar(xpos, revenue)

# xticks() function for replace names every bar in axis x go to company list.
# Example: in here replace names axis x xpos will replace go to company list

plt.xticks(xpos, company)
```

```python
# Give a y name, title and label on the graph/grafik (dibawah ini)
plt.bar(xpos, revenue) # code sebelumnya disalin

plt.ylabel("revenue(billion)")
plt.title("US Tech Stocks")
plt.bar(xpos, revenue, label="Revenue")
plt.xticks(xpos, company)

plt.legend() # klo memberikan label maka kasi fungsi ini biar keluar labelnya
```

```python

# add new tuple variable for add go to the graph/grafik.
profit = [40, 2, 32, 12]

plt.xticks(xpos, company)
# plt.bar(xpos, revenue)
plt.ylabel("revenue(billion)")
plt.title("US Tech Stocks")

plt.bar(xpos-0.2, revenue, label="Revenue")
plt.bar(xpos+0.2, profit, label="Profit") # this is code for create/make the bar_chart graph/grafik as new a tuple variable earlier/tadi and give a label for it
"""xpo-0.2 in first bar fucntion for apparently/seperintya shift/menggeser blue bar_chart go to left and xpos+0.2 in bar function second/kedua
for shift/menggeser bar_chart orange go to right"""

plt.legend()
```

```python
# add a width parameter for result over good or over neat/rapi
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2, profit, width=0.4, label="Profit")
plt.legend()
```

```python
# We can replace/changed the graph/grafik go to horizontal.
"""Explain: two barh() functions for replace the graph from vertical to horizontal and delete a width parameter and don't forget repalce xticks go to yticks and tish is thing
for move/memindahkan names company from x_axis to y_axis """

plt.yticks(xpos, company)
# plt.bar(xpos, revenue)
# plt.ylabel("revenue(billion)") tidak usah menggunakan code ini karena tidak cocok untuk penempatannya atau tidak akur atau tidak masuk akal
plt.title("US Tech Stocks")

plt.barh(xpos-0.2, revenue, label="Revenue")
plt.barh(xpos+0.2, profit, label="Profit")

plt.legend()
```
