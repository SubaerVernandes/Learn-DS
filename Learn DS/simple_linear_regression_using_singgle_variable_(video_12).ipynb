# First step

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.area, df.price)
```

# Creat graph over cool (modivied)

```python
%matplotlib inline
plt.xlabel("Area (sqr ft)")
plt.ylabel("Price (USD)")
plt.scatter(df.area, df.price, color="red", marker="+") # remember that marker style dot
```

# in move/below is does model

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price) # remember first parameter is fiture variable wheareas/sedangkan second parameter is label variable
```

# move/below 3 window code for search the variables in math formula

```python
reg.predict([[3300]]) # 3300 sesuai pertanyaan yg diinginkan diawal pertayanyaan dan ini juga sebagai variable x dalam rumus matematika diatas
# we can also use 5000 because this number is question above
```

```python
reg.coef_ # Reg.coef_ for search m variable value.
```

```python
reg.intercept_ #  reg.intercept_ for search b variable value.
```

# count/menghitung using math formula actually manual

image.png

```python
# m = 135.78767123, area = 3300, b =  180616.43835616432
price = 135.78767123 * 3300 + 180616.43835616432
price
```

# next step is plot

```python
# but on him github not use this code
%matplotlib inline
plt.xlabel("area", fontsize=20)
plt.ylabel("price", fontsize=20)
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
