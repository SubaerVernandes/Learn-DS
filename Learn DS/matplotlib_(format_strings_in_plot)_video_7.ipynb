```python
import matplotlib.pyplot as plt
%matplotlib inline
```

first, create the plot

```python
x = [1,2,3,4,5,6,7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(x,y)
```

```python
# Using "g+" parameter for changed/replace a grafik style, this is a grafik style for tell us about marker/penanda go on and down its grafik (titik² untuk penanda naik dan turunnya sebuah grafik)
plt.plot(x,y, 'g+')
```

https://matplotlib.org/api/as_gen/matplotlib.pyplot.plot.html

```python
# We also can replace/changed a grafik style, example here "--" sparateds/putus², "r" a red color, "*" for marker/penanda go on and down the grafik (these are three method 'all in one')
plt.plot(x,y, '--r*')
```

```python
# This is marker/penanda go on and down a grafik very cool
plt.plot(x,y, 'rD')
```

```python
# Can also replace/changed a grafik style with method step by step, code its before using three method all in one and here three mthod broken_down/dipecah bocome/menjadi three are parameter
plt.plot(x,y, color="red", marker="D", linestyle="")
```

```python
# The function markersize=20 for replace/changed of size the marker/penanda
plt.plot(x,y, color="red", marker="D", linestyle="", markersize=20)
```

```python
# an alpha=0.2 parameter for size level/tingkat bright/terang a grafik
plt.plot(x,y, color="green", alpha=0.2)
```
