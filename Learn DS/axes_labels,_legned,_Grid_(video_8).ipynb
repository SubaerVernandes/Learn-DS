```python
import matplotlib.pyplot as plt
# %matplotlib.inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 36, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
# days variable as axis/sumbu , max_t variable as axis/sumbu y, (we get information that in day hari ke-3 naik dan hari ke-5 turun (untuk max_t atau temperature maximum)
plt.plot(days, max_t)
```

```python
# We can also create a grafik all_in_one
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# 3 row pertama untuk memberikan keterangan
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

# 3 row kedua
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# Parameter label dan fungsi legend() untuk memberikan label pada grafik
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

# Parameter loc="upper right" untuk memindahkan posisi label ke pojok kanan atas
plt.legend(loc="upper right")

# Jika menggunakan parameter loc=best sama aja kegunaannya tanpa menabahkan parameter ke fungsi legend, yaitu sama2 menempatkan posisi label di pojok kiri bawah
```

```python
# Walaupun menggunakan fungsi untuk menempatkan posisi label dalam grafik ke pojok kiri bawah jika di varaible/kolom max_t di indeks ke-0 diganti ke 32, maka label dalam grafik akan berpindah sendiri ke tengah (tapi di kita engga ketengah labelnya)
days = [1,2,3,4,5,6,7]
max_t = [32, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 36, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]

# code dibawah hanya meng copy code sebelumnya
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

plt.legend(loc="best")

```

```python
# shadow=true parameter for give a shadow in label in the grafik (ini hanya untuk variasi, tapi ingat analisis kita yg lebih utama kolo urusan modivikasi belakangan, tapi klo di modiv kita terlihat terampil)

# code dibawah hanya meng copy code sebelumnya hanya untuk menambahkan parameter shadow=True pada fungsi lengend()
days = [1,2,3,4,5,6,7]
max_t = [32, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 36, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]


plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

# fontsize="samll" or fontsize="large" for replace/changed size label font (ini otomatis pindah kebawah klo menggunakan "large")
plt.legend(loc="best", shadow=True, fontsize="large")

# Give a grid() fucntion for ***memperjelas*** disumbu x atau hari keberapa yang naik ndan disumbu x atau hari keberapa yang turun (background kotak2)
plt.grid()
```
