```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/7_group_by/weather_by_cities.csv")
df
```

```python
g = df.groupby("city") # Mengelompokkan nilai berdasarkan citi (menemukan beberapa informasi)

# Menampilkan penegelompokan setelam dikelompokkan
for city, city_df in g:
  print(city)
  print(city_df)
```

#### Analogi

```python
# Menampilkan kelompok tertentu
g.get_group("mumbai")
```

```python
g.max() # melihat nilai maksimum di setiap kota berdasarkan pengelompokan
```
