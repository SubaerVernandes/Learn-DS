```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
# Create a new dataframe wich different
us_weather = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})
us_weather
```

analogi menggabungakan data frame

image.png

```python
# Menggabungkan data frame
df = pd.concat([india_weather, us_weather])
df
```

kita melihat bahwa data frame baru kita angka barisnya masih mengguanakan data sebelumnya, hal ini perlu diurutkan biar engga bingung

```python
# gunakan parameter ignore_index=True untuk memberikan angka baris yg berurutan
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
# setelah mengabungkan berikan label pada setiap data frame, hal ini biar engga tercampur semua (hal ini opsional sepertinya), label sesuai nama data setnya
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
# Untuk menampilkan label tertentu dai dataframe yang telah dilabelkan
df.loc["us"]
```

# two window as following for create new data frame

```python
# Create new dataframe
temperature_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangrole"],
    "temperature": [32, 45, 30],
    })
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "windspeed": [7, 12, 9],
    })
windspeed_df
```

```python
# Menggabungkan 2 dataframe yg baru dibuat
df = pd.concat([temperature_df, windspeed_df])
df
```

we can see the concat function documentation this link: https://pandas.pydata.org/docs/reference/api/pandas.concat.html

```python
# Menambahkan parameter axis=1 untuk menggabungkan data frame secara horizontal
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```

# 2 window dibawah untuk merubah nomor baris dari setiap data frame, tetapi yg diubah nomor barisnya adalah hanya data frame yg ke-2, nomor baris data frame yg ke-2 mengacu pada data frame yg pertama


explain: menambahkan parameter index untuk membuat nomor disetiap bais, index[1,0] bisa saja mengubah nomor setiap baris sesuai keinginan. Yang disini kita ingin mengubah nomor baris "dehli" ke-0 dan nomor baris "mumbai" ke-1 pada data frame yg kedua (pengangtian nomor baris ini mengacu ke nilai "dehli" dan nilai "mumbai" pada data frame pertama), tetapi posisi "delhi" tetap baris pertama hanya saja nomer barisnya dirubah dan posisi "mumbai" tetap baris kedua hanya saja nomor barisnya dirubah



langkah-langkah untuk metode ini:

1) tambahkan parameter index untuk data frame pertama

2) tambahkan parameter index pada data frame ke-2 yg ingin disesuaikan ke data frame yg pertama

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangrole"],
    "temperature": [32, 45, 30],
    }, index=[0,1,2])
temperature_df
```

```python
# Mengubah nomer baris sesuai data frame pertama atau sesuai keinginan
windspeed_df = pd.DataFrame({
    "city": ["mumbai", "delhi"],
    "windspeed": [7, 12],
    }, index=[1,0])
windspeed_df
```

```python
# Ingat parameter axis=1 untuk membuat data frame secara horizontal
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```
